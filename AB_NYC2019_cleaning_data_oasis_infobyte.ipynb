```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
NYC.columns = [col.lower() for col in NYC.columns]
```

```python
NYC['last_review'] = pd.to_datetime(NYC['last_review'])
```

```python
numeric_columns = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
for col in numeric_columns:
  sns.boxplot(x=NYC[col])
  plt.show()
```
